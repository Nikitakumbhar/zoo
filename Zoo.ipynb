```python

from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Label Encoding
z_df['class_type'].replace([1,2,3,4,5,6,7], [0,1,2,3,4,5,6], inplace=True)
z_df['species'].replace(['Mammal', 'Fish', 'Bird', 'Invertebrate', 'Bug', 'Amphibian','Reptile'],[0,1,2,3,4,5,6],inplace=True)
z_df.head()
```

```python
# split data into X and y
X = z_df[['species', 'hair', 'feathers', 'eggs',
       'milk', 'airborne', 'aquatic', 'predator', 'toothed', 'backbone',
       'breathes', 'venomous', 'fins', 'legs', 'tail', 'domestic', 'catsize']]
Y = z_df[ 'class_type']
```

```python
# split data into train and test sets
seed = 1
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

```python
# fit model no training data
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
print(model)
```

```python
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
y_pred
```

```python
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

**Train-Test split will be 80-20 ratio and random_state is to be set to 1,then overall accuracy rate of the model is 100%.**
